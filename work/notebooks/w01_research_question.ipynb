## 1. My Lane: Lane 2 (Refresh / Content Opportunity Scoring)
I chose this lane because for any organization, it is critical to identify which high-value content is losing visibility and needs to be updated to remain competitive. By prioritizing pages experiencing traffic decay, we can help the editorial team focus their limited resources on the content that will yield the highest return on investment.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd
df = pd.read_csv('flyrank-ml-internship/data/raw/content_refresh_anonymized.csv')

# Decay rate showing what percentage of the total pages are getting decayed
down_trend = df[df['trend_direction'] == 'down']
total_pages = df.shape[0]
down_pages = down_trend.shape[0]
decay_rate = (down_pages / total_pages) * 100
print(decay_rate)

# visibility impact shows how much total impressions are at stake if pages are not updated rightfully
visibility_impact = down_trend['impressions_90d'].sum()
print(visibility_impact)

#average lifespan for the decaying page
average_lifespan = down_trend['content_age_days'].mean()
print(average_lifespan)
```
